```python
#Remove as colunas ID
iris.drop('Id',axis=1,inplace=True)
```

## KNN (K-Nearest Neighbors)

Primeiro, padronizando as variáveis

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(iris.drop('Species',axis=1))
```

```python
scaled_features = scaler.transform(iris.drop('Species',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=iris.columns[:-1])
df_feat.head()
```

### Dividindo o dataset entre treino e teste

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_features,iris['Species'],test_size=0.30)
```

### Ajustando o modelo

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

### Previsão e avaliação

```python
pred = knn.predict(x_test)
```

```python
#Matriz de confusão
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

No primeiro teste o modelo teve 98% de precisão utilizanbdo apenas 1 vizinho

### Melhorando o modelo
Vou utilizar o método do cotovelo para escolher um bom valor K

```python
taxa_erros = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    taxa_erros.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,40),taxa_erros,color='blue',linestyle='dashed',marker='o',
        markerfacecolor='red',markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

```python
# PRIMEIRA COMPARAÇÃO RÁPIDA PARA O NOSSO ORIGINAL K = 20
knn = KNeighborsClassifier(n_neighbors=18)
knn.fit(x_train,y_train)
pred = knn.predict(x_test)
print('WITH = K=30')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
